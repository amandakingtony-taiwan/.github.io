```
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.dml.color import RGBColor
from pptx.enum.text import PP_ALIGN

def create_presentation():
    # 建立簡報物件
    prs = Presentation()

    # 定義輔助函式：新增標題與內容頁面
    def add_slide(prs, title_text, content_text_list):
        slide_layout = prs.slide_layouts[1] # 1 是標題+內容版面
        slide = prs.slides.add_slide(slide_layout)

        # 設定標題
        title = slide.shapes.title
        title.text = title_text
        title.text_frame.paragraphs[0].font.size = Pt(32)
        title.text_frame.paragraphs[0].font.bold = True
        title.text_frame.paragraphs[0].font.color.rgb = RGBColor(0, 51, 102) # 深藍色

        # 設定內容
        body_shape = slide.placeholders[1]
        tf = body_shape.text_frame
        tf.clear() # 清除預設空的段落

        for point in content_text_list:
            p = tf.add_paragraph()
            p.text = point
            p.font.size = Pt(20)
            p.space_after = Pt(14)
            # 如果是子項目 (以 '-' 開頭)，縮排
            if point.startswith("  -"):
                p.level = 1
                p.text = point.replace("  -", "").strip()

    # --- Slide 1: 封面 ---
    slide_layout = prs.slide_layouts[0] # 0 是封面版面
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    subtitle = slide.placeholders[1]

    title.text = "捷克手工具市場戰略分析\nKing Tony 品牌定位與機遇"
    title.text_frame.paragraphs[0].font.color.rgb = RGBColor(0, 51, 102)
    subtitle.text = "致：高階管理團隊 (Executive Team)\n日期：2026年1月22日"

    # --- Slide 2: 執行摘要 ---
    add_slide(prs, "1. 執行摘要 (Executive Summary)", [
        "市場地位：King Tony 已建立「專業技師理性首選」地位，位於中高階甜蜜點。",
        "競爭態勢：",
        "  - 頂層 (德系)：價格昂貴，僅工業採購。",
        "  - 底層 (Yato)：價格戰激烈，利潤低。",
        "  - 中層 (核心戰場)：主要機會在於承接 Tona Expert 流失客群。",
        "關鍵威脅：KS Tools (德系) 以「特種工具」與「高規低價」強勢切入。",
        "戰略目標：鞏固台灣製造信任感，優化產品組合以防禦 KS Tools。"
    ])

    # --- Slide 3: 市場定位與競爭版圖 ---
    add_slide(prs, "2. 市場定位：金字塔結構分析", [
        "頂層 (Premium)：Hazet, Stahlwille",
        "  - 特徵：工業/航太級，價格極高。",
        "  - King Tony 角色：高性價比替代方案 (價格約 60-70%)。",
        "中高層 (Sweet Spot)：King Tony vs. Tona vs. KS Tools",
        "  - King Tony 優勢：MIT 工業級品質信任、完整在地售後。",
        "  - Tona Expert：捷克國寶品牌，但因生產外移光環褪色。",
        "底層 (Budget)：Yato, Extol",
        "  - 特徵：價格侵略性強。",
        "  - 策略：避免價格肉搏，強調耐用度差異。"
    ])

    # --- Slide 4: 關鍵對手洞察：KS Tools ---
    add_slide(prs, "3. 關鍵威脅：KS Tools 的成功模式", [
        "威脅等級：高 (High)",
        "產品策略：",
        "  - 專注「解決方案型工具」(如引擎正時、噴油嘴工具)，非僅標準扳手。",
        "  - 明星商品「179件組」：無英制規格、含E-Torx，成為歐系車維修標配。",
        "通路策略：",
        "  - 數位能見度極高 (Alza, Mall 電商鋪貨完整)。",
        "  - 善用論壇口碑行銷，反應速度快。",
        "啟示：King Tony 需加速「特種工具」與「歐系專用套組」的開發。"
    ])

    # --- Slide 5: 客戶之聲 (Voice of Customer) ---
    add_slide(prs, "4. 市場洞察：來自捷克論壇的聲音", [
        "正面評價：",
        "  - 形象：Nezničitelné (堅不可摧/用不壞)。",
        "  - 權威：扭力扳手附校正證書，信任度極高。",
        "  - 耐用：氣動套筒 (Impact Sockets) 是工廠首選。",
        "負面痛點：",
        "  - 規格配置：部分套組仍含歐洲少用的英制規格 (Inch sizes)。",
        "  - 行銷包裝：相較對手，被認為包裝較「傳統樸實」，缺乏驚豔感。"
    ])

    # --- Slide 6: 戰略行動建議 ---
    add_slide(prs, "5. 戰略行動建議 (Action Plan)", [
        "產品優化 (Product)：",
        "  - 開發「捷克/歐系專用版」套組：移除英制，標配 E-Torx 與 VDE 工具。",
        "  - 搶灘電動車市場：加速引進 EV 維修專用絕緣工具。",
        "品牌溝通 (Brand)：",
        "  - 鎖定 Tona Expert 轉移戶：訴求「不變的工業品質」。",
        "  - 強化 MIT 標籤：區隔 Yato 等波蘭/中國品牌。",
        "通路深化 (Channel)：",
        "  - 發揮 Frýdek-Místek 總代理優勢，強調在地校正與零配件服務。"
    ])

    # --- Slide 7: 結語 ---
    add_slide(prs, "6. 結語：勝利方程式", [
        "King Tony 不需要做「最便宜」，也不需做「最奢侈」。",
        "",
        "我們的核心競爭力：",
        "維持工業級的「硬」品質 (Trust)",
        "+",
        "學習對手的「軟」身段 (產品組合靈活性)",
        "",
        "目標：成為捷克專業技師心中「唯一理性的選擇」。"
    ])

    # 存檔
    file_path = "King_Tony_Czech_Market_Strategy.pptx"
    prs.save(file_path)
    print(f"簡報已生成：{file_path}")

if __name__ == "__main__":
    create_presentation()
```
